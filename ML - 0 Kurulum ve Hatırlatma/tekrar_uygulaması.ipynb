# **Eksiye Dönük Bir Tekrar Çalışması (Retrospektif)**

Eksiye dönük bir tekrar çalışması, geçmişteki hataları ve olumsuz sonuçları analiz ederek, bu hatalardan öğrenme ve gelecekte daha iyi kararlar alabilme amacıyla gerçekleştirilen bir tür incelemedir. Bu tür çalışmalar, geçmişteki hataları tekrarlamamak ve daha verimli bir şekilde ilerlemek için önemlidir.

## Adımlar

1. **Projenin Geçmişi ve Amaçları**:
   - Projeyi neden başlattınız?
   - Hangi hedeflere ulaşmayı amaçladınız?
   - Hangi kararlar alındı?

2. **Hataların Tanımlanması**:
   - Projedeki hataları ve olumsuz sonuçları tanımlayın.
   - Hatalar nelerdir?

3. **Hataların Nedenleri**:
   - Her hatanın nedenlerini anlamaya çalışın.
   - Hataların nedenleri nelerdir?

4. **Öğrenme Fırsatları**:
   - Her hatadan ne öğrenebileceğinizi düşünün.
   - Hatalardan alınan dersler nelerdir?

5. **Düzeltici Eylemler**:
   - Hataları düzeltmek ve tekrarlanmasını önlemek için ne tür düzeltici eylemler alabilirsiniz?
   - Düzeltici eylemleri belirleyin ve uygulamak için bir plan yapın.

6. **Geleceğe Bakış**:
   - Projeyi nasıl geliştirebileceğinizi ve gelecekte daha başarılı olabileceğinizi düşünün.
   - Gelecekteki projelerinizde ne tür iyileştirmeler yapabilirsiniz?

7. **Yeniden Değerlendirme**:
   - Yaptığınız düzeltici eylemlerin etkisini düzenli olarak değerlendirin.
   - Projeyi izlemeye devam edin.

Eksiye dönük bir tekrar çalışması, bir organizasyonun veya bireyin sürekli olarak iyileşme ve büyüme fırsatlarını yakalamasına yardımcı olabilir. Hatalardan ders almak ve bu dersleri gelecekteki projelerde uygulamak, başarı şansını artırabilir ve daha iyi sonuçlar elde etmenize katkıda bulunabilir.

## **Sütunları Seçme**

Birden fazla sütunu seçmek için `.loc[]` veya `.iloc[]` kullanabilirsiniz. İşte birden fazla sütunu seçmek için bu indeksleme yöntemlerini kullanmanın iki farklı yolu:

### İsimle Sütun Seçimi (`.loc[]`):

```python
sepal_lengths = df['SepalLengthCm']
petal_lengths = df['PetalLengthCm']
```

```python
sepal_lengths = df.loc[:, 'SepalLengthCm']  # 'SepalLengthCm' sütunu
petal_lengths = df.loc[:, 'PetalLengthCm']  # 'PetalLengthCm' sütunu
```

#### Birden fazla sütun seçme

```python
secilen_sutunlar = df.loc[:, ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
```

```python
secilen_sutunlar
```

### Pozisyonla Sütun Seçimi (`.iloc[]`):

```python
sepal_lengths = df.iloc[:, 0]  # İlk sütun, SepalLengthCm
petal_lengths = df.iloc[:, 2]  # Üçüncü sütun, PetalLengthCm
```

#### Birden fazla sütun seçme

```python
secilen_sutunlar = df.iloc[:, 0:4]
```

```python
secilen_sutunlar
```

## **Satırları Seçme**

### İsimle Satır Seçimi (`.loc[]`):

```python
iris_setosa = df.loc[df['Tür'] == 'Iris-setosa']
iris_versicolor = df.loc[df['Tür'] == 'Iris-versicolor']
iris_virginica = df.loc[df['Tür'] == 'Iris-virginica']
```

```python
iris_setosa.head()
```

```python
iris_versicolor.head()
```

```python
iris_virginica.head()
```

### Pozisyonla Satır Seçimi (`.iloc[]`):

```python
iris_setosa = df.iloc[:50, :]  # 0 ile 50. satıra kadar olan tüm satırlar seçilir (50. satır dahil değil).
iris_versicolor = df.iloc[50:100, :]  # 50 ile 100. satıra kadar olan tüm satırlar seçilir (100. satır dahil değil).
iris_virginica = df.iloc[100:150, :]  # 100 ile 150. satıra kadar olan tüm satırlar seçilir
```

## **Histogram ve Pasta Grafiği Oluşturma**

```python
tur_dagilimi = df['Tür'].value_counts()

# Sayıları bar grafik olarak çizme
plt.figure(figsize=(15, 7))
tur_dagilimi.plot(kind='bar', color=['blue', 'orange', 'green'])
plt.xlabel('Türler')
plt.ylabel('Örnek Sayısı')
plt.title('Her Türün Sayısı')
plt.show()
```

```python
species_counts = df['Tür'].value_counts()

plt.figure(figsize=(15, 7))  # Grafiğin boyutunu belirleyebilirsiniz
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Eşit bir daire şekli oluşturmak için
plt.show()
```

```python
df_turler = pd.DataFrame(df.Tür.value_counts().reset_index().values, columns=["Tür", "count"])
df_turler.head()
```
