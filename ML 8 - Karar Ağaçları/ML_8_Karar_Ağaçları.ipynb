# **1. Karar Ağacı Algoritmasına Giriş** <a class="anchor" id="1"></a>

[İçindekiler](#0.1)

Karar Ağacı (Decision Tree) algoritması, en popüler makine öğrenmesi algoritmalarından biridir. Belirli bir sorunu çözmek için ağaç benzeri bir yapı ve bunların olası kombinasyonlarını kullanır. Hem sınıflandırma hem de regresyon amacıyla kullanılabilecek denetimli öğrenme algoritmaları sınıfına aittir.

Karar ağacı, kök düğüm (root node), dallar (branches) ve yaprak düğümleri (leaf node) içeren bir yapıdır. Her dahili düğüm, bir öznitelik üzerindeki bir testi belirtir, her dal bir testin sonucunu belirtir ve her yaprak düğüm ise bir sınıf etiketine sahiptir. Ağacın en üstteki düğümü kök düğümdür.

Karar Ağacı algoritmasını uygularken bazı varsayımlarda bulunuruz. Bunlar aşağıda listelenmiştir: -

1. Başlangıçta, tüm eğitim seti kök olarak kabul edilir.
2. Özniteliklerin değerleri kategorik olmalıdır. Değerler sürekli ise, model oluşturmadan önce ayrık hale çevrilmelidir.
3. Öznitelikleri ağacın kök veya iç düğümü olarak yerleştirme sırası, bazı istatistiksel yaklaşımlar kullanılarak yapılır.


# **2. Sınıflandırma ve Regresyon Ağaçları (CART)** <a class="anchor" id="2"></a>

[İçindekiler](#0.1)


Günümüzde Karar Ağacı algoritması, **Sınıflandırma ve Regresyon Ağaçları** (Classification and Regression Trees) anlamına gelen modern adı **CART** ile bilinmektedir. Sınıflandırma ve Regresyon Ağaçları veya **CART**, Leo Breiman tarafından sınıflandırma ve regresyon modelleme problemleri için kullanılabilen Karar Ağacı algoritmalarına atıfta bulunmak için kullanılan bir terimdir.





# **3. Karar Ağacı Algoritması Terminolojisi** <a class="anchor" id="3"></a>

[İçindekiler](#0.1)


- Bir Karar Ağacı algoritmasında, her bir dahili düğümün bir öznitelik üzerindeki bir testi temsil ettiği, her dalın testin sonucunu temsil ettiği ve her yaprak düğümün bir sınıf etiketini temsil ettiği ağaç benzeri bir yapı vardır. Kök düğümden yaprak düğüme giden yollar, sınıflandırma kurallarını temsil eder.

- Karar Ağacı algoritmasında yer alan terimler aşağıdaki gibidir: -


## Kök Düğüm (Root Node)

- Tüm popülasyonu veya örneği temsil eder. Bu ayrıca iki veya daha fazla homojen kümeye bölünür.


## Dallanma (Splitting)

- Bir düğümü iki veya daha fazla alt düğüme bölme işlemidir.


## Karar Düğümü (Decision Node)

- Bir alt düğüm başka alt düğümlere bölündüğünde, buna karar düğümü denir.


## Yaprak / Terminal Düğümü (Leaf/Terminal Node)

- Bölünmeyen düğümlere Yaprak veya Terminal düğümleri denir.


## Budama (Pruning)

- Bir karar düğümünün alt düğümlerini kaldırdığımızda bu işleme budama denir. Bölmenin tam tersi bir süreçtir.


## Dal / Alt ağaç (Branch / Sub-tree)

- Tüm ağacın bir alt bölümüne dal veya alt ağaç denir.


## Üst ve Alt Düğüm (Parent and Child Node)

- Alt düğümlere bölünmüş bir düğümde, alt düğümlerin bir ana düğümün çocukları olduğu düşünülür.

# **4. Algoritma Nasıl Çalışıyor?** <a class="anchor" id="4"></a>

[İçindekiler](#0.1)

Karar Ağacı algoritması, hem sınıflandırma hem de regresyon görevleri için sık ve yaygın olarak kullanılan denetimli makine öğrenmesi algoritmalarından biridir. Karar Ağacı algoritmasının arkasındaki çalışma prensibini anlamak çok basittir.


Karar Ağacı algoritmasının çalışma prensibi aşağıdaki gibidir: -


1.	Veri kümesindeki her bir öznitelik için Karar Ağacı algoritması bir düğüm oluşturur. En önemli öznitelik kök düğüme yerleştirilir.

2.	Eldeki görevi değerlendirmek için, kök düğümden başlıyoruz, koşulumuza veya kararımıza uyan ilgili düğümü takip ederek ağaçta aşağı iniyoruz.

3.	Bu işlem bir yaprak düğüme ulaşılana kadar devam eder. Karar Ağacının tahminini veya sonucunu içerir.


# **5. Öznitelikler Nasıl Seçilir?** <a class="anchor" id="5"></a>

[İçindekiler](#0.1)

Karar Ağacı uygulamasındaki ilk zorluk, kök düğüm ve ağacın her düzeyinde üzerine çalışacağımız öznitelikleri belirlemektir. Bu süreç, **öznitelik seçimi** olarak bilinir. Her düzeyde kök düğüm olarak kabul edilebilecek özniteliği tanımlamak için farklı öznitelik seçim ölçütleri bulunmaktadır.

İki adet popüler öznitelik seçim ölçütü vardır. Bunlar aşağıdaki gibidir: -


- **Information gain  (Bilgi Kazanımı)**

- **Gini indeksi**

**Bilgi kazanımı** kriter olarak kullanılırken  özniteliklerin kategorik olduğu varsayılır. Şayet **Gini indeksi** kullanılırsa, özniteliklerin sürekli olduğu varsayılır. Bu öznitelik seçim ölçütleri aşağıda açıklanmıştır.


## **5.1 Bilgi Kazanımı** <a class="anchor" id="5.1"></a>

[İçindekiler](#0.1)


Bilgi kazanımını ölçüt olarak kullanarak her bir özniteliğin içerdiği bilgiyi tahmin etmeye çalışırız. Bilgi Kazanımı kavramını anlamak için **Entropi** adlı başka bir kavramı bilmemiz gerekir.

## **Entropi**

Entropi, verilen veri kümesindeki safsızlığı (saf/ yalın olmama - impurity) ölçer. Fizik ve Matematikte entropi, X rastgele değişkeninin rastgeleliği veya belirsizliği olarak adlandırılır. Bilgi teorisinde, bir grup örnekteki safsızlığı ifade eder. **Bilgi kazancı** entropideki azalmadır. Bilgi kazancı, verilen öznitelik değerlerine dayalı olarak, bölünmeden önceki entropi ile veri kümesinin bölünmesinden sonraki ortalama entropi arasındaki farkı hesaplar.

Entropi aşağıdaki formülle temsil edilir: -




![Entropy](http://www.learnbymarketing.com/wp-content/uploads/2016/02/entropy-formula.png)


Burada **c** veri setindeki sınıf sayısıdır ve **pi**  i.sınıfla ilişkili olasılıktır.

Karar Ağacı algoritması, bilgi kazancını hesaplamak için entropiyi kullanır. Böylece, her bir özelliğin **entropi ölçüsündeki** düşüşü hesaplayarak onların bilgi kazancını hesaplayabiliriz. En yüksek bilgi kazancına sahip öznitelik, düğümde bölme özniteliği olarak seçilir.

## **5.2 Gini indeks** <a class="anchor" id="5.2"></a>

[İçindekiler](#0.1)

**CART'ın (Kategorik ve Regresyon Ağaçları)** kullandığı başka bir öznitelik seçim ölçüsü **Gini indeksidir**. Bölünme noktaları oluşturmak için Gini yöntemini kullanır.


Gini indeksi aşağıdaki formül ile gösterilebilir: -

## **Gini index**

![Gini index](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRzYHkcmZKKp2sJN1HpHvw-NgqbD9EnapnbXozXRgajrSGvEnYy&s)

Burada yine **c** sınıf sayısıdır ve **pi**  i. sınıfla ilişkili olasılıktır.

Gini indeksi, bir popülasyondan rastgele iki öğe seçersek, aynı sınıfta olmaları gerektiğini ve popülasyon saf ise bunun için olasılık 1'dir demektedir.

Kategorik hedef değişken "Başarılı" veya "Başarısız" ile çalışır. Yalnızca ikili bölmeler gerçekleştirir. Gini değeri ne kadar yüksek olursa homojenlik de o kadar yüksek olur. CART (Sınıflandırma ve Regresyon Ağacı), ikili bölmeler oluşturmak için Gini yöntemini kullanır.

Bölünme için Gini Hesaplama Adımları

1.	Başarı ve başarısızlık olasılığının karesinin formül toplamını kullanarak alt düğümler için Gini'yi hesaplayın (p^2+q^2).

2.	Bölünmenin her bir düğümünün ağırlıklı Gini puanını kullanarak bölme için Gini'yi hesaplayın.


Ayrık değerli bir öznitelik durumunda, seçilen için minimum gini indeksini veren altküme, bir bölme özniteliği olarak seçilir. Sürekli değerli nitelikler söz konusu olduğunda, strateji, olası bir bölünme noktası olarak her bir bitişik değer çiftini ve bölme noktası olarak seçilen daha küçük gini indeksine sahip noktayı seçmektir. Minimum Gini indeksi olan nitelik, bölme niteliği olarak seçilir.

```python
# X_train içindeki veri tiplerini inceleyelim

X_train.dtypes
```

```python
encoder = ce.OrdinalEncoder(cols=['satis_fiyati', 'bakim_masrafi', 'kapi_sayisi', 'kisi_kapasitesi', 'bagaj_kapasitesi', 'guvenlik'])


X_train = encoder.fit_transform(X_train)

X_test = encoder.transform(X_test)
```

# **12. Gini İndeks Kriterini Kullanarak Karar Ağacı Sınıflandırması** <a class="anchor" id="12"></a>

[İçindekiler](#0.1)

```python
# import DecisionTreeClassifier

from sklearn.tree import DecisionTreeClassifier
```

```python
# Modeli Gini İndeksi ile başlatmak için uygun parametre girişini sağlayalım.

clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)


# Modeli eğitelim
clf_gini.fit(X_train, y_train)
```

### Sonuçları test verisi üzerinde kestirelim

```python
y_pred_gini = clf_gini.predict(X_test)
```

### Doğruluk puanını kontrol edelim

```python
from sklearn.metrics import accuracy_score

print('Gini İndeksi Kriteri Kullanılmış Modelin Doğruluk Puanı: {0:0.4f}'. format(accuracy_score(y_test, y_pred_gini)))
```

Burada, **y_test** gerçek sınıf etiketlerini, **y_pred_gini** ise test verileri üzerinden tahmin edilmiş sınıf etiketlerini temsil etmektedir.

### Eğitim ve Test Verilerinin Doğruluğunu Karşılaştıralım


```python
y_pred_train_gini = clf_gini.predict(X_train)

y_pred_train_gini
```

```python
print('Eğitim veri seti doğruluk puanı: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_gini)))
```

### Aşırı ya da Eksik Öğrenmenin Olup Olmadığını Kontrol Edelim

```python


print('Eğitim veri seti doğruluk puanı: {:.4f}'.format(clf_gini.score(X_train, y_train)))

print('Test veri seti doğruluk puanı: {:.4f}'.format(clf_gini.score(X_test, y_test)))
```

Burada eğitim seti doğruluk puanı 0.7865 iken, test seti doğruluğu 0.8021'dir. Bu iki değer oldukça yakındır. Yani, aşırı öğrenme belirtisi yok.


### Karar Ağaçlarını Görselleştirelim

```python
plt.figure(figsize=(12,8))

from sklearn import tree

tree.plot_tree(clf_gini.fit(X_train, y_train)) 
```

### graphviz Kütüphanesini kullanarak karar ağacını görselleştirelim

```python
import graphviz 
dot_data = tree.export_graphviz(clf_gini, out_file=None, 
                              feature_names=X_train.columns,  
                              class_names=y_train,  
                              filled=True, rounded=True,  
                              special_characters=True)

graph = graphviz.Source(dot_data) 

graph 
```
